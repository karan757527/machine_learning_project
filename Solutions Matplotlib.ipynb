```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import numpy as np
%matplotlib inline
import pandas as pd
df = pd.read_csv('company_sales_data.csv')
print("Data Set we will use is:")
df.head()
```

### 2

```python
def per(p):
    h=[]
    s=0
    for pa in p:
        x=pa.get_width()
        s+=x
        h.append(x)
    h=np.array(h)
    h=(h/s)*100
    return h
     

fig=plt.figure(figsize=(10,8))
people = ('G1','G2','G3','G4','G5','G6','G7','G8')
data=np.array([[ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477],
    [ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
    [ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
    [ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]])


color=['red','green','gold','cyan']
patch=[]
patch.append(plt.barh(people,data[0],color=color[0]))
x=np.zeros(8)
for i in range (1,4):
    x=x+data[i-1]
    patch.append(plt.barh(people,data[i],color=color[i],left=x))
plt.xlabel("Scores")


for p in patch:
    x=per(p)
    for i, v in enumerate(x):
        a=p[i].get_x()+p[i].get_width()/2
        plt.text(a-1, i, str(round(v, 2))+"%", va="center")
plt.show()
```

### 5

```python
plt.plot('month_number','total_profit',data=df,color='r')
plt.xlabel("Month Number")
plt.ylabel("Profit in Dollar")
plt.xticks(range(1,13))
plt.yticks(range(100000,500001,100000))
plt.title("Company profit per month")
plt.show()
```

### 6

```python
plt.plot('month_number','total_profit','r--o',data=df,linewidth=3,mfc="black")
plt.xlabel("Month Number")
plt.ylabel("Sold unit number")
plt.xticks(range(1,13))
plt.yticks(range(100000,500001,100000))
plt.title("Company Sales data of last year")
plt.legend(["Profit data of last year"],loc="lower right")
plt.show()
```

### 7

```python
col=list(df.columns)
col=col[1:7]
color=["blue","orange",'green','red','purple','brown']
for i in range(6):
    plt.plot('month_number',col[i],'-o',color=color[i],data=df,linewidth=3,mfc=color[i])
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.xticks(range(1,13))
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.title("Sales data")
plt.legend([x.title()+" Sales Data" for x in col])
plt.show()
```
