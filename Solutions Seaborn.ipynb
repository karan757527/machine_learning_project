## Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
```

## Import Datasets

```python
df=sns.load_dataset('titanic')
df1=sns.load_dataset('diamonds')
df2=sns.load_dataset('tips')
df3=sns.load_dataset('fmri')
df4=sns.load_dataset('iris')
df5=sns.load_dataset('exercise')
```

## 1

```python
sns.distplot(df['fare'],kde=False,color='r',hist_kws={'ec':'r'},bins=30)
```

## 2

```python
sns.boxplot(y=df['age'],x=df['class'])
```

## 3

```python
sns.countplot(df['sex'])
```

## 4

```python
g=sns.PairGrid(df,y_vars='survived',x_vars=['class','sex','who','alone'],size=5,aspect=.5)
g.map(sns.pointplot)
```

## 5

```python
sns.set_style("white")
sns.barplot(df1['cut'],df1['price'],hue=df1['color'])
plt.legend(loc=1)
```

## 6

```python
sns.countplot(df1['cut'],color='r')
```

## 7

```python
sns.distplot(df1['carat'],kde=False,hist_kws={'width':0.11})
plt.yticks(range(0,12001,2000))
```

## 8

```python
sns.set_style("whitegrid")
sns.violinplot(df2['day'],df2['total_bill'],hue=df2['sex'],split=True,inner="quartile")
```

## 9

```python
sns.barplot(x="class", y="survived", hue="sex", data=df)
```

## 10

```python
sns.set_style("darkgrid")
sns.lineplot(x="timepoint", y="signal",hue="region", style="event", data=df3)
```

## 11

```python
sns.set_style("white")
sns.scatterplot(df4['sepal_length'],df4['sepal_width'],hue=df4['species'])
```

## 12

```python
setosa = df4.loc[df4.species == "setosa"]
virginica = df4.loc[df4.species == "virginica"]
versicolor = df4.loc[df4.species == "versicolor"]
sns.kdeplot(setosa['petal_length'],label='setosa')
sns.kdeplot(virginica['petal_length'],label='virignica')
sns.kdeplot(versicolor['petal_length'],label='versicolor')
```

## 13

```python
sns.jointplot(y="sepal_width", x="sepal_length", data=df4,marginal_kws=dict(bins=15, rug=True),annot_kws=dict(stat="r"),edgecolor="w", linewidth=1)
```

## 14

```python
sns.pairplot(df4, hue="species")
```

## 15

```python
sns.set_style("whitegrid")
sns.catplot(x="time", y="pulse",hue="kind", col="diet",data=df5, kind="point",palette=sns.light_palette("blue", reverse=True),capsize=.2);
```
